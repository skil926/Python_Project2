# An Exploration of Pubilc Health Data

## Import Libraries

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib. pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Read Data

```python
# Open csv file and import data
df = pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv')
df
```

```python
df.describe()
```

## Cleaning Data

```python
# Remove unknown data - consider data with "?" as unknown data - removing columns with unknown data
df.dropna(axis=1)
```

```python
# futher filtered the data to keep just the columns we want to analyze and call the dataset df2
df2= df[['encounter_id', 'patient_nbr', 'gender', 'age', 'time_in_hospital', 'max_glu_serum', 'A1Cresult','change', 'diabetesMed', 'readmitted','insulin']]
df2
```

## Descriptive Statistics

```python
# count of patients by gender and age - we see highest count in the age range 70-80 in both males and females
df2.groupby((['gender','age']))['patient_nbr'].count().sort_values(ascending = False)
```

```python
# count of patients who were prescribed diabetic medication by gender and age
#the same group - patients between 70-80 stay the highest group of patients even when filtered for those who do not take diabetic medication. 
df3=df2[df2['diabetesMed']=='No']
df3.groupby((['gender','age']))['time_in_hospital'].mean().sort_values(ascending = False)
```

```python
# average time spent in hospital by gender and age - we see the highest average time of stay in the hospital in the 80-100 age range in both males and females
df2.groupby((['gender','age']))['time_in_hospital'].mean().sort_values(ascending = False)
```

## Visualization

```python
# plot to compare 
ax=sns.boxplot(x='age', y='time_in_hospital', hue='gender', data=df2)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

## Conclusion

### There seems to be a direct relationship between time spent in the hospital and patient's age range regardless of gender. Although there are 2 noticeable differences where gender seems to play a role in the time spent in the hospital. Males tend to spend more time in the hospital at age 0-10 (active males) while females seem to spend more time during the age 50-60 (menopause).
